```python

    Corpus - Body of text, singular. Corpora is the plural of this. Example: A collection of medical journals.
    Lexicon - Words and their meanings. Example: English dictionary. Consider, however, that various fields 
        will have different lexicons. For example: To a financial investor, the first meaning for the word "Bull" 
            is someone who is confident about the market, as compared to the common English lexicon, where the first 
            meaning for the word "Bull" is an animal. As such, there is a special lexicon for financial investors, 
            doctors, children, mechanics, and so on.
    Token - Each "entity" that is a part of whatever was split up based on rules. For examples, each word is a token 
    when a sentence is "tokenized" into words. Each sentence can also be a token, if you tokenized the sentences out 
    of a paragraph.
```

```python
Generally, computers use numbers for everything, but we often see directly in programming where we use binary signals 
(True or False, which directly translate to 1 or 0, which originates directly from either the presence of an electrical 
 signal (True, 1), or not (False, 0)). To do this, we need a way to convert words to values, in numbers, or signal 
patterns. The process of converting data to something a computer can understand is referred to as "pre-processing." 
One of the major forms of pre-processing is going to be filtering out useless data. In natural language processing, 
useless words (data), are referred to as stop words.
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

word_tokenize = word_tokenize(content)
print(len(word_tokenize))
stopword = set(stopwords.words('english'))

final_words = []
for i in word_tokenize:
    if i not in stopword:
        final_words.append(i)
        
print(final_words)
print(len(final_words))

print("\n\n")
print(len(word_tokenize)-len(final_words))
```

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
train_text = state_union.raw("/home/venkatesh/Documents/Resume-Mining---FInal-Year-Project/Tutorials/Infy - CPDD.txt")
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
```

```python
tokenized = custom_sent_tokenizer.tokenize(train_text)
```

```python
def process_content():
    try:
        for i in tokenized[:5]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)

    except Exception as e:
        print(str(e))


process_content()
```
